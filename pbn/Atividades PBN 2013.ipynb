### Função para atualizar o estado pela matriz `mi` em um passo

```python
def update_state(state, mi):
    def x_t(v, i):
        if v > 0:
            return 1
        elif v < 0:
            return 0
        else:
            return state[i]
            
    i = mi.dot(state)
    return map(lambda t: x_t(t[1], t[0]), enumerate(i))
    
    
```

```python
# All states possible for the four genes expression
all_states = [np.array([g0, g1, g2, g3]) for g0 in xrange(0,2) for g1 in xrange(0,2) for g2 in xrange(0,2) for g3 in xrange(0,2)]


# Function to convert from binary list to integer number
def to_num(bl): return sum(map(lambda t: t[1]*2**(t[0]), enumerate(reversed(bl))))
```

## Função que acha os atratores, seus tamanhos e suas bacias de atração

```python
def find_atts(all_states, matrix):
    atts = {}
    
    for i,s in enumerate(all_states):
        ls = [to_num(s)]
        s0 = s
        while True:
            sn = update_state(s0, matrix)
            sn_num = to_num(sn)
            if sn_num in ls:
                if not atts.has_key(sn_num):
                    atts[sn_num] = {}
                    atts[sn_num]['basin_size'] = 1
                    atts[sn_num]['size'] = len(ls) - ls.index(sn_num)
                else:
                    atts[sn_num]['basin_size'] += 1                    
                break
            else:
                ls.append(to_num(sn))
                s0 = sn
    
    return atts
```

```python
atts = { "Matrix {0:02d}".format(m[0]): find_atts(all_states, m[1]) for m in enumerate(ms)}

for m in atts.keys():
    display(HTML("<b>" + m + "</b>"))
    print "Número de atratores: {}".format(len(atts[m].keys()))
    for a in atts[m].keys():
        print "{}\t=> Tamanho: {},\t Tamanho da bacia: {}".format(a, atts[m][a]['size'], atts[m][a]['basin_size'])
```

```python
def state_graph(all_states, matrix):
    size = len(all_states)
    adj_matrix = np.zeros(shape=(size, size))
    
    for i,s in enumerate(all_states):
        ls = [to_num(s)]
        s0 = s
        while True:
            sn = update_state(s0, matrix)
            s0_num = to_num(s0)
            sn_num = to_num(sn)
            
            adj_matrix[s0_num][sn_num] = 1
            
            if sn_num in ls:              
                break
            else:
                ls.append(to_num(sn))
                s0 = sn
    
    return adj_matrix
```

```python
for i, m in enumerate(ms):
    display(HTML("<h3> Matrix {0:02d} </h3>".format(i)))    
    sg = state_graph(all_states, m).tolist()
    graph = pydot.graph_from_adjacency_matrix(sg, node_prefix=-1, directed=True)

    graph_filename = 'state-{0:02d}.png'.format(i)
    f = open(graph_filename, 'wb')
    f.write(graph.create_png())
    f.close()
    
    display(Image(graph_filename))
```
