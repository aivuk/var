```python
import numpy as np
from sklearn.decomposition import PCA, KernelPCA
import pylab as pl
```

```python
X = np.loadtxt('/home/walrus/ratos-2.csv', delimiter=',')
```

```python
kpca = KernelPCA(kernel="precomputed", gamma=10)
```

```python
X_kpca = kpca.fit_transform(X)
X_back = kpca.inverse_transform(X_kpca)
pca = PCA()
X_pca = pca.fit_transform(X)
```

```python
len(X_kpca[X_kpca[:,1] > 0.5])
```

```python
pl.plot(X_kpca[:, 0], X_kpca[:, 1], '.')
pl.title("Projection by KPCA")
pl.xlabel("1st principal component in space induced by $\phi$")
pl.ylabel("2nd component")

pl.show()
```

```python
pl.plot(X_pca[:, 0], X_pca[:, 1], "ro")
pl.title("Projection by PCA")
pl.xlabel("1st principal component")
pl.ylabel("2nd component")
```
