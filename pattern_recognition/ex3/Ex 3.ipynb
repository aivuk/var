```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.neighbors import NearestNeighbors
import scipy.spatial.distance as distance
import scipy.cluster.hierarchy as sch
import matplotlib.gridspec as gridspec
```

# Carrega os dados e remove valores com mais que dois NaN's

Primeiro carregamos os dados de um arquivo csv, ignorando linhas em branco:

```python
data = pd.read_csv('http://www.molbiolcell.org/content/suppl/2003/09/24/9.12.3273.DC1/CDCDATA.txt', sep='\t', skip_blank_lines=True)
```

Depois removemos todas as linhas (genes) que possuem mais de dois NaN (Not a Number). Pra isso usamos os filtros do próprio pandas para gerar um novo data frame:

```python
data_clean = data[data.isnull().sum(axis=1) <= 2]
```

E removemos com o comando abaixo a primeira coluna que possui apenas os nomes dos genes, obtendo finalmente um DataFrame possuindo apenas números que representam as expressões do genes em diferentes condições experimentais (colunas).

```python
numeric_data = data_clean.ix[:, 1:]
```

# Substitui valores NaN pelo método KNNimputation

Antes de analisarmos os dados iremos substituir os valores NaN pelo método KNN imputation. Em tal método primeiro encontramos dado um gene que possua um (ou dois) NaN em uma (ou duas) coluna(s), quais genes que possuem tais colunas definidas que estão mais próximos deste gene. Feito isso fazemos uma média dos genes dos vizinhos com pesos proporcionais a similaridade do gene com colunas
faltantes (NaN's).

Para realizar este passo faremos duas funções, uma `neighbors_from` que retorna quais genes possuem expressões mais similares a um gene dado, fornecendo a linha que o gene está como argumento.

A outra função `impute_in` substitui os valores NaN num experimento de um dado gene pela média ponderada das expressões dos 15 vizinhos mais próximos desse gene.

Para calcular os vizinhos mais próximos eu considerei apenas os genes que não possuiam nenhum NaN. Isso porque mesmo se um gene possuisse um valor em um experimento onde o gene que se está procurando os vizinhos possuise NaN, caso ele tivesse um NaN em outro experimento a distância desconsiderando esse outro NaN não necessariamente representaria uma distância correta.

```python
def neighbors_from(numeric_data, row):
    ''' Função que dado uma matriz de dados `numeric_data` e uma linha `row`, retorna os 15 vizinhos
        mais próximos da linha especificada.
    '''
    # Primeiro dado uma linha (gene) da nossa matriz de dados, encontramos todas
    # colunas (experimentos) que não possuem NaN.
    not_null_in_row = numeric_data.irow(row)[numeric_data.irow(row).notnull()]
    # Concatenamos essa linha que agora não possui NaNs, com todas as linhas que não possuem
    # nenhum NaN. Dessas linhas pegamos apenas as colunas iguais as colunas do gene que estamos
    # interessados em encontrar os vizinhos mais próximos através da indexação [not_null_in_row.index].
    d = pd.concat([not_null_in_row.to_frame().T, numeric_data[numeric_data.isnull().sum(axis=1) == 0][not_null_in_row.index]])
    # Feito isso encontramos os vizinhos mais próximos do conjunto inteiro.
    nbrs = NearestNeighbors(n_neighbors=15, algorithm='ball_tree').fit(d)
    # E retornamos os índices dos vizinhos e as distâncias do gene que estamos interessados, que
    # é a primeira linha do novo dataframe `d` que criamos.
    return nbrs.kneighbors(d.irow(0))

def impute_in(numeric_data, row):
    ''' Função que dado uma matriz de dados `numeric_data` e uma linha `row`, substitui os valores NaN's
        na linha por uma média ponderada dos vizinhos mais próximos da linha que possuem todos valores definidos.
    '''
    # Primeiro criamos uma cópia da linha que estamos interessados.
    nc = numeric_data.irow(row).copy()
    # Calculamos as distâncias que serão utilizadas para calcular os pesos e os índices dos vizinhos mais próximos.
    weights, to_input = neighbors_from(numeric_data, row)
    # Convertemos as distâncias para similaridades.
    weights = 1.0/weights[0][1:].reshape(len(weights[0]) - 1,1)
    # Pegamos apenas os índices das linhas diferentes do índice da própria linha que estamos realizando a substituição dos NaN.
    to_input = to_input[0][1:]
    # E usamos as 'mágicas' de broadcast e vetorização do Numpy/Pandas para substituir em uma linha os valores NaN's.
    # Note que já estamos normalizando os pesos e multiplicando linhas a linha dos dados sem NaN, apenas das colunas que
    # estamos interessados, ou seja, as colunas que são NaN na linha `row` que passamos como argumento pra função.
    nc[numeric_data.irow(row).isnull()] = (weights * (numeric_data.iloc[to_input].ix[:,numeric_data.irow(row).isnull()])).sum()/weights.sum()
    return nc
```

```python
%time new_data = pd.DataFrame([impute_in(numeric_data, row) for row in range(numeric_data.shape[0])])
```

Finalmente normalizamos os dados para possuirem média 0 e desvio padrão 1:

```python
scaled_data = preprocessing.scale(new_data)
```

# Calcula matrizes de distância

Calculamos as matrizes de distâncias para os genes `pairwise_dists` e para os experimentos `pairwise_dists_columns` usando o módulo distance do Scipy:

```python
pairwise_dists = distance.squareform(distance.pdist(scaled_data,metric='correlation'))
pairwise_dists_columns = distance.squareform(distance.pdist(scaled_data.T,metric='correlation'))
```

# Encontra os clusters

Finalmente utilizamos as informações das distâncias entre os genes para encontrar os diferentes cluster (e sub clusters) cada gene pertence:

```python
clusters = sch.linkage(pairwise_dists, method='average')
```

É possível notar visualmente que as linhas agrupadas nos mesmo clusters possuem um padrão de expressão semelhantes ao longo dos diferentes experimentos (colunas do gráfico acima).

# Encontrando os clusters para uma dada distância mínima

Utilizamos as informações sobre os clustes encontrados para achar qual cluster cada gene pertence. Para definir o grau de agrupamento que estamos interessados, passamos a distância mínima 15 como argumento para a função `fcluster`.

```python
cs = sch.fcluster(clusters, 15, criterion='distance')
```

```python
plt.figure(figsize=(10, 8))
gg=sch.dendrogram(clusters, p=int(cs.max()), truncate_mode='lastp', orientation='right')
```
