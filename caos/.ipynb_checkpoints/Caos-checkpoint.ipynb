```python
def mk_f(a, n=1):
    def f(x):
        return a*x*(1 - x)
    return f
```

```python
f = mk_f(2.707)
```

```python
x = linspace(0,1, num=100)
```

```python
plot(x, f(x))
#plot(x[50:70], (lambda y: f(0.7) + df(0.7)*(y - 0.7))(x[50:70]))
plot(x, mk_f(3.414)(x))
plot(x, x)
```

```python
import functools
```

```python
def iter_f(f, n):
    def fn(x):
        for i in xrange(n):
            x = f(x)
        return x
    return fn
```

```python
plot(x, iter_f(mk_f(3.7), 3)(x))
plot(x, iter_f(mk_f(3.9), 3)(x), '--')
plot(x, x)
```

```python
def deriv(f, h):
    def df(x):
        return (f(x + h) - f(x))/h
    return df
```

```python
df = deriv(f, 0.1)
```

```python
def logistic_map(a_min=0, a_max=4, n_a=1000, n_iter=300, x0=0.5, discard_points=200):
    for i in range(n_a):
        a = a_min + (float(i)/float(n_a))*(a_max - a_min)
        f_points = [iter_f(mk_f(a), i)(x0) for i in range(1, n_iter)][discard_points:]
        rcParams['figure.figsize']  = 15, 13
        rcParams['figure.dpi']  = 300
        scatter([a for _ in range(len(f_points))], f_points, s=1, facecolor='0', lw = 0)
```

```python
logistic_map(a_min=3.82, a_max=3.86)
```
